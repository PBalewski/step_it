```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

def imshow(img):
    vmin = 0
    vmax = img.max()
    if vmax > 1:
        vmax = 255
    if vmax > 1:
        img = img.astype(np.uint8)
    img = np.array(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    plt.axis('off')
    if len(img.shape) == 2:
        plt.imshow(img, cmap='gray', vmin=vmin, vmax=vmax)
    else:
        plt.imshow(img, vmin=vmin, vmax=vmax)
    plt.show()
```

```python
image_path = 'Materials_13.12/cat.jpeg'
cat_face_cascade = cv2.CascadeClassifier('Materials_13.12/haarcascade_frontalcatface_extended.xml')
image = cv2.imread(image_path)
cat_face = cat_face_cascade.detectMultiScale(image)
for (x,y,w,h) in cat_face:
    cv2.rectangle(image, (x, y),(x+w, y+h), (0,0,255), 3)
print(cat_face)

imshow(image)
print(image.shape)
```

```python
import cv2
from PIL import Image

image_path = 'Materials_13.12/cat.jpeg'
cat_face_cascade = cv2.CascadeClassifier('Materials_13.12/haarcascade_frontalcatface_extended.xml')
image = cv2.imread(image_path)
cat_face = cat_face_cascade.detectMultiScale(image)

cat = Image.open(image_path)
glasses = Image.open('Materials_13.12/glasses.png')
cat = cat.convert("RGBA")
glasses = glasses.convert("RGBA")
for (x,y,w,h) in cat_face:
    glasses = glasses.resize((w, int(h/3)))
cat.paste(glasses, (x, int(y+h/4)), glasses)
cat.save("cat_with_glasses.png")
cat_with_glasses = cv2.imread("cat_with_glasses.png")
imshow(cat_with_glasses)
```
